## Giving context to AI
lets get technical on 'how we contextually assign/ask ai'
in this example i will try do demonstrate how we create very simple R.A.G

In our first code lets demonstrate our problem:
the function generate_answer here is very basics
its only takes query and search if any of strings in documents have matched string

```python
import numpy as np

# Step 1: Small dataset
documents = [
    "Book: The Great Gatsby, Genre : Art",
    "Book: To Kill a Mockingbird, Genre : Literature",
    "Book: 1984, Genre : Horror",
    "Book: Ghost in the machine, Genre : Philosphy"
]

def generate_answer(query, doc):
    for doc in documents:
        if query.lower() in doc.lower():
            return doc
    return "I don't know."

query = "Horror"
answer = generate_answer("Horror", documents)
print(f"Question: {query}")
print(f"Answer: {answer}")

query = "which book type has genre: horror"
answer = generate_answer("query", documents)
print(f"Question: {query}")
print(f"Answer: {answer}")
```

As we can see here its very primitive retrival here, 
although we gives query "which book type has genre: horror"
the sistem cant answer this question because its only matches string in query
with the documents. 

Lets see how we can improve it using embeeding technique.

the idea is we have to measure the similiarty between the 'questions' and the sentence in documents
so, we use math and statistic here. the simplest idea is using histogram that counted word in the documents

lets see how we calculated histogram here, below is the function to generate histogram for each sentence in documents

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Small dataset
documents = [
    "Book: The Great Gatsby , Genre : Fiction",
    "Book: To Kill a Mockingbird , Genre : Literature",
    "Book: 1984 , Genre : Dystopian Fiction",
    "Book: Ghost in the machine , Genre : Philosphy"
]

# Vocabulary used for embedding
vocab = ['book', 'genre', 'fiction', 'dystopian', 'great', 'gatsby', 'mockingbird,', '1984', 'literature', 'philosphy', 'mockingbird']

# Step 2: Simple embedding function (word frequency)
def create_embedding(text):
    words = text.lower().split()
    vector = np.zeros(len(vocab))
    for i, word in enumerate(vocab):
        vector[i] = words.count(word)
    norm = np.linalg.norm(vector) or 1  # Avoid division by zero
    return vector / norm

# Step 3: Create embeddings
doc_vectors = [create_embedding(doc) for doc in documents]

# Step 4: Plot histograms with sentence as title
plt.figure(figsize=(12, 8))

for i, (vector, doc) in enumerate(zip(doc_vectors, documents)):
    plt.subplot(len(doc_vectors), 1, i + 1)
    plt.bar(vocab, vector)
    plt.title(f"Document {i + 1}: {doc}", fontsize=10)
    plt.ylabel("Normalized Frequency")
    plt.ylim(0, 1)
    plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

so we have a histogram here that count words in vocab

vocab = ['book', 'genre', 'fiction', 'dystopian', 'great', 'gatsby', 'mockingbird', '1984']

the next step is to create same histogram for the questions

```python
query = "what genre is Kill a Mockingbird"
query_vector = create_embedding(query)
plt.figure(figsize=(12, 8))
plt.bar(vocab, query_vector)
plt.title(f"Query: {query}", fontsize=10)
plt.ylabel("Normalized Frequency")
plt.ylim(0, 1)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

Now, we have histogram of words in question
and also
we have histograms, for each sentence in documents

lets use these histogram to get similiarty.
the problem that we want to solve is: 
which sentences in documents, that might have same context with the questions

we wil use a measure called: cosine_similiarity
implementation of code are:

```python
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2) or 1)
```

this function calculates similiarity between questions and sentences in documents
 lets try it:

```python
similarities = [cosine_similarity(query_vector, doc_vec) for doc_vec in doc_vectors]
print(f"similarities: {similarities}")
plt.figure(figsize=(12, 8))
plt.bar(['1', '2', '3', '4'], similarities)
plt.title(f"Query x Sentences score", fontsize=10)
plt.ylim(0, 1)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
```

this diagram shows what is the 'score' of similiarty between questions and the sentences
lets recall our mission:

the sentences in our documents are
1. "Book: The Great Gatsby , Genre : Fiction",
2. "Book: To Kill a Mockingbird , Genre : Literature",
3. "Book: 1984 , Genre : Dystopian Fiction",
4. "Book: Ghost in the machine , Genre : Philosphy"

the question are:
"what genre is Kill a Mockingbird"

so, in the plot above document number 2 has the highest score for similiarity
hence, we will select document number 2 

```python
relevant_doc_index = np.argmax(similarities)
relevant_doc = documents[relevant_doc_index]
print(f"relevant_docs are: {relevant_doc}")
```

## Advancement

so now we will advance our excersise using existing tools that available
i will be using SentenceTransformer


```python
from sentence_transformers import SentenceTransformer, util

documents = [
    "Book: The Great Gatsby , Genre : Fiction",
    "Book: To Kill a Mockingbird , Genre : Literature",
    "Book: 1984 , Genre : Dystopian Fiction",
    "Book: Ghost in the machine , Genre : Philosphy"
]

retriever_model = SentenceTransformer('all-MiniLM-L6-v2')
note_embeddings = retriever_model.encode(documents, convert_to_tensor=True)
print(f"note_embeddings:{note_embeddings}")
```

```python
question = "what genre is Kill a Mockingbird"
question_embedding = retriever_model.encode(question, convert_to_tensor=True)
print(f"note_embeddings:{question_embedding}")
```

```python
import torch
from sentence_transformers import util

similarities = util.cos_sim(question_embedding, note_embeddings)[0]
print(f"similarities:{similarities}")
most_relevant_idx = torch.argmax(similarities).item()
retrieved_note = documents[most_relevant_idx]
print(f"Retrieved Note: {retrieved_note}")
print(f"Similarity Scores: {similarities}")
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Encode notes and question
embeddings = retriever_model.encode(documents + [question], convert_to_numpy=True)

# Reduce to 2D using PCA
pca = PCA(n_components=2)
embeddings_2d = pca.fit_transform(embeddings)

# Plot
plt.figure(figsize=(8, 6))
for i, text in enumerate(documents + [question]):
    plt.scatter(embeddings_2d[i, 0], embeddings_2d[i, 1], label=text[:20] + "...")
    plt.text(embeddings_2d[i, 0] + 0.01, embeddings_2d[i, 1], f"{'Note' if i < 3 else 'Question'} {i+1 if i < 3 else ''}")
plt.legend()
plt.title("2D Visualization of Embeddings")
plt.show()
```

```python
question = "What genre book has titled Ghost?"
question_embedding = retriever_model.encode(question, convert_to_tensor=True)
similarities = util.cos_sim(question_embedding, note_embeddings)[0]
most_relevant_idx = torch.argmax(similarities).item()
print(f"Retrieved Note: {documents[most_relevant_idx]}")
print(f"Similarity Scores: {similarities}")
```

lets advance the experiments

we will learn about Attention Transformer here:


```python
from transformers import AutoTokenizer, AutoModel, pipeline

def retrieve_relevant_text(question, embeddings, texts):
    # Encode the question
    question_embedding = retriever_model.encode(question, convert_to_tensor=True)
    # Compute cosine similarity
    similarities = util.cos_sim(question_embedding, embeddings)[0]
    # Find the most relevant text
    most_relevant_idx = torch.argmax(similarities).item()
    return texts[most_relevant_idx], similarities

structured_embeddings = retriever_model.encode(documents, convert_to_tensor=True)
question = "What genre book has titled Ghost?"
retrieved_note, structured_similarities = retrieve_relevant_text(question, structured_embeddings, documents)
print("\nStructured Data Retrieval:")
print(f"Question: {question}")
print(f"Retrieved Note: {retrieved_note}")
print(f"Similarity Scores: {structured_similarities.cpu().numpy()}")

model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

inputs = tokenizer(retrieved_note, return_tensors="pt", truncation=True)

with torch.no_grad():
    outputs = model(**inputs, output_attentions=True)
attention = outputs.attentions[-1][0].mean(dim=0)  # Last layer, averaged across heads
```
